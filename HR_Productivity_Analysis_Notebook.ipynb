# Monthly Reports

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Date to datetime
daily['Date'] = pd.to_datetime(daily['Date'])

daily.set_index('Date', inplace = True)

```

```python
monthly = daily.groupby(['Employee','Department']).resample('M').mean()
```

```python
monthly = monthly.reset_index()
```

```python
min_date = daily['Date'].min().date()
max_date = daily['Date'].max().date()
```

```python
monthly['Productive Total'] = monthly['Productive Active Hours'] + monthly['Productive Passive Hours']
```

```python
# Remove the last 3 characters from the 'Employee_YearMonth' column
monthly['chart_date'] = monthly['Date'].str.slice(0, -3)
```

```python
monthly
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define the department names
departments = monthly['Department'].unique()

# Loop through each department
for department in departments:
    # Filter the data for the current department
    df_department = monthly[monthly['Department'] == department]

    # Get unique employees within the department
    employees = df_department['Employee'].unique()

    # Define the number of rows and columns for subplots
    n = len(employees)
    ncols = 2
    nrows = n // ncols + n % ncols
    pos = range(1, n + 1)

    # Create a new figure for each department
    fig = plt.figure(figsize=(20, 10*nrows))

    # Create subplots for each employee
    for p, employee in zip(pos, employees):
        # Filter the data for the current employee
        df_employee = df_department[df_department['Employee'] == employee]

        # Select only the last 5 months for this employee
        df_employee = df_employee.tail(10)

        # Sort the DataFrame by Date
        df_employee = df_employee.sort_values('Date')

        # Create a subplot in the right position
        ax = fig.add_subplot(nrows, ncols, p)

        # Create a bar plot with stacked=True for the Productive Active and Passive hours. Add `legend=False` to avoid generating legends here.
        df_employee.set_index('chart_date')[['Productive Active Hours', 'Productive Passive Hours']].plot(kind='bar', stacked=True, ax=ax, legend=False)

        # Create line plots for Productive Goal, Upper Threshold, and Lower Threshold
        ax.axhline(y=6, color='g', linestyle='-')  # Assuming the Productive Goal is constant and equals to 6
        ax.axhline(y=7.8, color='b', linestyle='-')  # Assuming the Upper Threshold is constant and equals to 7.8
        ax.axhline(y=4.2, color='r', linestyle='-')  # Assuming the Lower Threshold is constant and equals to 4.2

        # Set the title and labels
        ax.set_title(f'Employee: {employee}')
        ax.set_xlabel('Date')
        ax.set_ylabel('Hours')

        # Rotate x-axis labels for better readability
        plt.xticks(rotation=0)

    # Set a main title for the figure
    fig.suptitle(f'Department: {department}', fontsize=16)

    # Create legend entries manually
    active_hours_patch = mpatches.Patch(color='blue', label='Productive Active Hours')  # Assuming 'Productive Active Hours' are plotted in blue
    passive_hours_patch = mpatches.Patch(color='orange', label='Productive Passive Hours')  # Assuming 'Productive Passive Hours' are plotted in orange
    productive_goal_line = plt.Line2D([0], [0], color='g', linewidth=2, label='Productive Goal')
    upper_threshold_line = plt.Line2D([0], [0], color='b', linewidth=2, label='Upper Threshold')
    lower_threshold_line = plt.Line2D([0], [0], color='r', linewidth=2, label='Lower Threshold')

    # Add legend to the figure, this time positioned at the top center
    fig.legend(handles=[active_hours_patch, passive_hours_patch, productive_goal_line, upper_threshold_line, lower_threshold_line],
               loc='upper center', bbox_to_anchor=(0.5, 0.96), ncol=5)

    # Improve layout
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)

    # Save the figure to a file
    fig.savefig(f'monthly_{department}_plot_{min_date} - {max_date}.png')

# Show the last figure
plt.show()
```

# Weekly Reports

```python
from matplotlib.colors import ListedColormap, BoundaryNorm

# Set the index to 'Date' and 'Employee'
weekly.set_index(['Date', 'Employee'], inplace=True)

# Get the unique departments
departments = weekly['Department'].unique()

# Define the color map
cmap = ListedColormap(['yellow', 'green', 'red'])
norm = BoundaryNorm([0, 4.2, 7.8, weekly['Productive Total'].max()], cmap.N)

# Loop over the departments
for department in departments:
    # Create a sub-dataframe for the current department
    df_dept = weekly[weekly['Department'] == department]

    # Unstack the 'Employee' level of the index and select the 'Productive Total' column
    df_total_dept = df_dept['Productive Total'].unstack('Employee')

    # Plot the heatmap
    plt.figure(figsize=(10,10))
    sns.heatmap(df_total_dept, cmap=cmap, norm=norm, linewidths=.5)
    plt.title(f'Weekly Productive Total Hours for {department} Department')
    plt.show()
```
